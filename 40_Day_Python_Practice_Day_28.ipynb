```python
text = "Cng c Suno AI nhanh chng nhn c s ch  t ngi dng khi c th to bi ht ch vi vi cu lnh. Phin bn mi nht V3 Alpha mi c gii thiu cui thng 2, c bn min ph vi 10 bi ht mi ngy."
```

```python
from pprint import pprint
```

```python
from translate import Translator

translator_en = Translator(to_lang='en', from_lang='vi')
translation_en = translator_en.translate(text)

translator_jp = Translator(to_lang='ja', from_lang='vi')
translation_jp = translator_jp.translate(text)

pprint(translation_en)
pprint(translation_jp)
```

```python
from googletrans import Translator

gg_translator = Translator()
gg_translation_en = gg_translator.translate(text, dest='en')
gg_translation_jp = gg_translator.translate(text, dest='ja')

pprint(gg_translation_en.text)
pprint(gg_translation_jp.text)
```
