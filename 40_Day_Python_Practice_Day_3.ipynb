# Giải phương trình bậc 2

```python
def quadratic_equation(a, b, c):
    x1 = 0
    x2 = 0

    delta = b ** 2 - 4 * a * c

    if a == 0:
        if b == 0 and c == 0:
            print('phương trình có vô số nghiệm')
        elif b == 0 and c != 0:
            print('phương trình vô nghiệm')
        elif b !=0:
            x = -b / c
            print(f'x = {x}')
    else:
        if delta > 0:
            x1 = (-b - delta ** 0.5) / (2 * a)
            x2 = (-b + delta ** 0.5) / (2 * a)
            print(f'x1 = {x1} và x2 = {x2}')
        elif delta == 0:
            x1 = x2 = -b / (2 * a)
            print(f'x1 = x2 = {x1}')
        else:
            print('phương trình vô nghiệm')
```

```python
quadratic_equation(a=2, b=6, c=4)
```

```python
quadratic_equation(a=1, b=2, c=1)
```

```python
quadratic_equation(a=4, b=6, c=3)
```
