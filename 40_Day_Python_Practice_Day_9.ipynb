# Câu 2

```python
lst_data = sorted(lst_data)

n = len(lst_data)

if n % 2 == 0:
    median = (lst_data[int(n / 2)] + lst_data[int((n - 1)/2)]) / 2
else:
    median = lst_data[n] / 2

print(f"Median: {median}")
```
