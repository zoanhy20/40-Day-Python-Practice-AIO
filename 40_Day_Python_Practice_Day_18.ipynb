## list comprehension

```python
stop_words = ['I', 'love', 'and', 'to']

input = 'I love AI and listen to music'

result1 = [word for word in input.split() if word not in stop_words]

result1
```

## for

```python
stop_words = ['I', 'love', 'and', 'to']

input = 'I love AI and listen to music'

result2 = []

for word in input.split():
    if word not in stop_words:
        result2.append(word)

result2
```
