```python
np.random.seed(2024)

arr_random = np.random.rand(3, 4)
arr_int = np.random.randint(-10, 10, (3, 4))

arr_random, arr_int
```

```python
tf.random.set_seed(2024)

tf_random = tf.random.uniform((3, 4))
tf_int = tf.random.uniform((3, 4), -10, 10, dtype=tf.dtypes.int32)

tf_random, tf_int
```

```python
torch.manual_seed(2024)

torch_random = torch.rand((3, 4))
torch_int = torch.randint(-10, 10, (3, 4))

torch_random, torch_int
```
