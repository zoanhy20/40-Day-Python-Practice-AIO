```python
corpus = ['Ti thch mn Ton', 'Ti thch AI', 'Ti thch m nhc']

vocab = []

for doc in corpus:
    doc = doc.lower().split()

    for word in doc:
        if word not in vocab:
            vocab.append(word)

print(vocab)
```

```python
sample = 'Ti thch AI thch Ton'

vector = [0] * len(vocab)

for word in sample.lower().split():
    if word in vocab:
        idx = vocab.index(word)
        vector[idx] += 1

print(vector)
```
