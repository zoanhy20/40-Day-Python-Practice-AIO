```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Câu 1

```python
img1 = cv2.imread('1.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

plt.axis('off')
plt.imshow(img1)
plt.show()
```

# Câu 2

```python
img2 = cv2.imread('1.jpg', 0)

plt.axis('off')
plt.imshow(img2, cmap='gray')
plt.show()
```

# Câu 3

```python
img3 = cv2.imread('1.jpg')
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)

increase_img3 = img3.astype(np.float32) + 50
increase_img3 = np.clip(increase_img3, 0, 255)
increase_img3 = increase_img3.astype(np.uint8)

decrease_img3 = img3.astype(np.float32) - 80
decrease_img3 = np.clip(increase_img3, 0, 255)
decrease_img3 = decrease_img3.astype(np.uint8)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img3)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(increase_img3)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(decrease_img3)
plt.axis('off')

plt.show()
```
