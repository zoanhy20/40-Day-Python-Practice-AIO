# Cu 1

```python
import sqlite3

connection = sqlite3.connect('database.sqlite')
cursor = connection.cursor()
```

```python
cursor.execute("""
CREATE TABLE PRODUCT (
    ID INT PRIMARY KEY,
    NAME TEXT NOT NULL,
    PRICE INT NOT NULL
)
""")
```

```python
cursor.execute("""
INSERT INTO PRODUCT(ID, NAME, PRICE)
VALUES
    ('1', 'iPhone 15', '18000000'),
    ('2', 'Galaxy Z-Fold 5', '30000000')
""")
```

```python
connection.commit()

import pandas as pd

data = pd.read_sql_query('SELECT * FROM PRODUCT', connection)
print(data)
```

# Cu 2

```python
cursor.execute("""
UPDATE PRODUCT
SET PRICE = '50000000'
WHERE ID = 2
""")
```

```python
connection.commit()

data = pd.read_sql_query('SELECT * FROM PRODUCT', connection)
print(data)
```
