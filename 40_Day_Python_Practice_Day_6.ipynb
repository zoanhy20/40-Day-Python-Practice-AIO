# Method 1: if / else

```python
time = input('Input time: ')

if time == '6 AM':
    print('Wake up')
elif time == '7 AM':
    print('Gym')
elif time == '8 AM':
    print('Work')
else:
    print('Do something else')
```

## Method 2: dictionary

```python
todo = {
    '6 AM': 'Wake up',
    '7 AM': 'Gym',
    '8 AM': 'Work',
}

time = input('Input time: ')

todo.get(time, 'Do something else')
```

# Method 3: match case

```python
time = input('Input time: ')

match time:
    case '6 AM':
        print('Wake up')
    case '7 AM':
        print('Gym')
    case '8 AM':
        print('Work')
    case _:
        print('Do something else')
```
