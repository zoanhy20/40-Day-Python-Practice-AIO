# Cch 1: S dng if/else

```python
def calculate_can_chi_calendar(year):
    result = ''

    result_can = ''

    if year % 10 == 0:
        result_can = 'Canh'
    elif year % 10 == 1:
        result_can = 'Tn'
    elif year % 10 == 2:
        result_can = 'Nhm'
    elif year % 10 == 3:
        result_can = 'Qu'
    elif year % 10 == 4:
        result_can = 'Gip'
    elif year % 10 == 5:
        result_can = 'p'
    elif year % 10 == 6:
        result_can = 'Bnh'
    elif year % 10 == 7:
        result_can = 'inh'
    elif year % 10 == 8:
        result_can = 'Mu'
    elif year % 10 == 9:
        result_can = 'K'

    result_chi = ''

    if year % 12 == 0:
        result_chi = 'Thn'
    elif year % 12 == 1:
        result_chi = 'Du'
    elif year % 12 == 2:
        result_chi = 'Tut'
    elif year % 12 == 3:
        result_chi = 'Hi'
    elif year % 12 == 4:
        result_chi = 'T'
    elif year % 12 == 5:
        result_chi = 'Su'
    elif year % 12 == 6:
        result_chi = 'Dn'
    elif year % 12 == 7:
        result_chi = 'Mo'
    elif year % 12 == 8:
        result_chi = 'Thn'
    elif year % 12 == 9:
        result_chi = 'T'
    elif year % 12 == 10:
        result_chi = 'Ng'
    elif year % 12 == 11:
        result_chi = 'Mi'

    result = result_can + ' ' + result_chi

    return result
```

```python
calculate_can_chi_calendar(2024)
```

```python
calculate_can_chi_calendar(2023)
```

# Cch 2: S dng list

```python
lst_can = ['Canh', 'Tn', 'Nhm', 'Qu', 'Gip', 'p', 'Bnh', 'inh', 'Mu', 'K']
lst_chi = ['Thn', 'Du', 'Tut', 'Hi', 'T', 'Su', 'Dn', 'Mo', 'Thn', 'T', 'Ng', 'Mi']

def calculate_can_chi_calendar(year):
    result = ''

    id_can = year % 10
    id_chi = year % 12

    result = lst_can[id_can] + ' ' + lst_chi[id_chi]

    return result
```

```python
calculate_can_chi_calendar(2024)
```

```python
calculate_can_chi_calendar(2023)
```
