# Cách 1: Sử dụng if/else

```python
def calculate_can_chi_calendar(year):
    result = ''

    result_can = ''

    if year % 10 == 0:
        result_can = 'Canh'
    elif year % 10 == 1:
        result_can = 'Tân'
    elif year % 10 == 2:
        result_can = 'Nhâm'
    elif year % 10 == 3:
        result_can = 'Quý'
    elif year % 10 == 4:
        result_can = 'Giáp'
    elif year % 10 == 5:
        result_can = 'Ấp'
    elif year % 10 == 6:
        result_can = 'Bính'
    elif year % 10 == 7:
        result_can = 'Đinh'
    elif year % 10 == 8:
        result_can = 'Mậu'
    elif year % 10 == 9:
        result_can = 'Kỷ'

    result_chi = ''

    if year % 12 == 0:
        result_chi = 'Thân'
    elif year % 12 == 1:
        result_chi = 'Dậu'
    elif year % 12 == 2:
        result_chi = 'Tuất'
    elif year % 12 == 3:
        result_chi = 'Hợi'
    elif year % 12 == 4:
        result_chi = 'Tý'
    elif year % 12 == 5:
        result_chi = 'Sửu'
    elif year % 12 == 6:
        result_chi = 'Dần'
    elif year % 12 == 7:
        result_chi = 'Mẹo'
    elif year % 12 == 8:
        result_chi = 'Thìn'
    elif year % 12 == 9:
        result_chi = 'Tỵ'
    elif year % 12 == 10:
        result_chi = 'Ngọ'
    elif year % 12 == 11:
        result_chi = 'Mùi'

    result = result_can + ' ' + result_chi

    return result
```

```python
calculate_can_chi_calendar(2024)
```

```python
calculate_can_chi_calendar(2023)
```

# Cách 2: Sử dụng list

```python
lst_can = ['Canh', 'Tân', 'Nhâm', 'Quý', 'Giáp', 'Ấp', 'Bính', 'Đinh', 'Mậu', 'Kỷ']
lst_chi = ['Thân', 'Dậu', 'Tuất', 'Hợi', 'Tý', 'Sửu', 'Dần', 'Mẹo', 'Thìn', 'Tỵ', 'Ngọ', 'Mùi']

def calculate_can_chi_calendar(year):
    result = ''

    id_can = year % 10
    id_chi = year % 12

    result = lst_can[id_can] + ' ' + lst_chi[id_chi]

    return result
```

```python
calculate_can_chi_calendar(2024)
```

```python
calculate_can_chi_calendar(2023)
```
