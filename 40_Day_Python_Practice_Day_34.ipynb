```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('2.jpg', 0)

lst_sigma = [2.5, 5.0, 10.0]

fig, ax = plt.subplots(2, 2)

def gaussian_kernel(size, sigma):
    if size % 2 == 0:
        size += 1

    max_point = size // 2
    min_point = -max_point

    K = np.zeros((size, size))
    for x in range(min_point, max_point + 1):
        for y in range(min_point, max_point + 1):
            value = (1 / (2 * np.pi * sigma ** 2)) * np.exp(-(x ** 2 + y ** 2) / (2 * sigma ** 2))
            K[x - min_point, y - min_point] = value

    return K

def display(ax, row, col, img, title):
    ax[row, col].imshow(img, cmap='gray')
    ax[row, col].set_title(title)
    ax[row, col].axis('off')


idx = 0
for row in range(2):
    for col in range(2):
        if row == 0 and col == 0:
            display(ax, row, col, img, 'Origin Image')

        else:
            kernel = gaussian_kernel(5, lst_sigma[idx])
            img_gaussian= cv2.filter2D(img, -1, kernel)
            display(ax, row, col, img_gaussian, f'Sigma = {lst_sigma[idx]}')
            idx += 1

plt.tight_layout()

plt.show()
```
