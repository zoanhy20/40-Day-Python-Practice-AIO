```python
import numpy as np
import tensorflow as tf
import torch
```

# Cu 1

```python
arr1 = list(range(1, 6))

print(np.array(arr1))
print(tf.constant(arr1))
print(torch.tensor(arr1))
```

# Cu 2

```python
def print_info(arr_lib, flag=0):
    print(arr_lib, arr_lib.shape, arr_lib.dtype, type(arr_lib), arr_lib.device if flag == 1 else '', end='\n\n')
```

```python
arr2 = [list(range(1, 6)), list(range(6, 11))]

arr2_np = np.array(arr2)
arr2_tf = tf.constant(arr2)
arr2_torch = torch.tensor(arr2)

print('Numpy: ')
print_info(arr2_np)
print('Tensorflow: ')
print_info(arr2_tf, flag=1)
print('Torch: ')
print_info(arr2_torch, flag=1)
```
