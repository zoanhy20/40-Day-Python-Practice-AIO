```python
import numpy as np
import tensorflow as tf
import torch
```

# Câu 1

```python
min_val = -10
max_val = 10
```

```python
np.random.seed(2024)

arr_1 = np.random.randint(min_val, max_val, size=(3, 4))
arr_2 = np.random.randint(min_val, max_val, size=(3, 4))

print(arr_1)
print(arr_2, arr_2.T)

print(arr_1 @ arr_2.T)
```

```python
tf.random.set_seed(2024)

tf_1 = tf.random.uniform((3, 4), min_val, max_val, dtype=tf.dtypes.int32)
tf_2 = tf.random.uniform((3, 4), min_val, max_val, dtype=tf.dtypes.int32)

print(tf_1)
print(tf_2, tf.transpose(tf_2))

print(tf.matmul(tf_1, tf.transpose(tf_2)))
```

```python
torch.manual_seed(2024)

torch_1 = torch.randint(min_val, max_val, (3, 4))
torch_2 = torch.randint(min_val, max_val, (3, 4))

print(torch_1)
print(torch_2, torch.t(torch_2))

print(torch.matmul(torch_1, torch.t(torch_2)))
```

# Câu 2

```python
np.random.seed(2024)

arr = np.random.randint(min_val, max_val, size=(3, 3))

total_sum_arr = np.sum(arr)
column_sum_arr = np.sum(arr, 0)
row_sum_arr = np.sum(arr, 1)

print(total_sum_arr, column_sum_arr, row_sum_arr)
```

```python
tf.random.set_seed(2024)

tf_init = tf.random.uniform((3, 3), min_val, max_val, dtype=tf.dtypes.int32)

total_sum_tf = tf.reduce_sum(tf_init)
column_sum_tf = tf.reduce_sum(tf_init, axis=0)
row_sum_tf = tf.reduce_sum(tf_init, axis=1)

print(total_sum_tf, column_sum_tf, row_sum_tf)
```

```python
torch.manual_seed(2024)

torch_init = torch.randint(min_val, max_val, (3, 3))

total_sum_torch = torch.sum(torch_init)
column_sum_torch = torch.sum(torch_init, 0)
row_sum_torch = torch.sum(torch_init, 1)

print(total_sum_torch, column_sum_torch, row_sum_torch)
```
