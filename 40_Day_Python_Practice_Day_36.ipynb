```python
import sqlite3

connection = sqlite3.connect('database.sqlite')
cursor = connection.cursor()
```

```python
cursor.execute("""

""")
```

```python
cursor.execute("""
CREATE TABLE STOCK (
    ID INT PRIMARY KEY,
    NAME TEXT NOT NULL,
    BUY INT NOT NULL,
    INVESTOR TEXT NOT NULL
)
""")
```

```python
cursor.execute("""
INSERT INTO STOCK(ID, NAME, BUY, INVESTOR)
VALUES
    ('1', 'ACB', '29.45', 'Nguyen'),
    ('2', 'VIC','44.55','Nguyen'),
    ('3', 'GMD','74.30','Nguyen'),
    ('4', 'ACB','28.45','Vinh'),
    ('5', 'VIC','40.55','Vinh'),
    ('6', 'GMD','60.30','Vinh')
""")
```

```python
connection.commit()

import pandas as pd

data = pd.read_sql_query('SELECT * FROM STOCK', connection)
print(data)
```

# Cu 1

```python
query = """
SELECT SUM(BUY) AS total_buy
FROM STOCK
"""

sum_buy = pd.read_sql_query(query, connection)
print(f"Tng gi bn: {sum_buy['total_buy']}")
```

# Cu 2

```python
query = """
SELECT INVESTOR, NAME, MAX(BUY) AS PRICE
FROM STOCK
GROUP BY INVESTOR
"""

data = pd.read_sql_query(query, connection)
print(data)
```
