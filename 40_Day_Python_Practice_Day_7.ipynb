# Câu 1

```python
lst_data = list(range(2, 13, 2))
print(lst_data)
```

# Câu 2

```python
for value in lst_data:
    if value % 3 == 0:
        lst_data.remove(value)

print(lst_data)
```

# Câu 3

```python
for i in range(1, 4):
    lst_data.append(i)

for i in range(6, 9):
    lst_data.insert(3 + i - 5, i)

print(lst_data)
```

# Câu 4

```python
lst_data = [0 if v % 2 == 0 or v % 5 == 0 else v  for v in lst_data]

print(lst_data)
```
