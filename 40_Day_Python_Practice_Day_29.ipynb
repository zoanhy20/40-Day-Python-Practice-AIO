```python
import librosa

import numpy as np
import matplotlib.pyplot as plt

audio, sr = librosa.load('data.wav')
time = np.linspace(0, audio.shape[0] / sr, num=audio.shape[0])

plt.plot(time, audio, color='blue')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (quantized)')
plt.title('Wav file visualization')
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```

```python
import speech_recognition as sr

r = sr.Recognizer()
audio_filename = 'data.wav'

my_audio = sr.AudioFile(audio_filename)

with my_audio as source:
    audio = r.record(source)

print(type(audio))
your_speech = r.recognize_google(audio, language='en')
print('Audio transcription: ', your_speech)
```

```python
from gtts import gTTS
import librosa
import numpy as np

lang='vi'
output_filename = './record.mp3'
content = """
Lm li t nhng th n gin nht, chuyn tr nh nh lc mi yu
Tnh mnh nh th nc hoa phng pht, nng nn nh bt bn riu
Ny ngi yu du nu trong mt chc bun su eo bm chng ta
Th ng vi trch c sao tn khc v nim au y s qua
"""
output = gTTS(content, lang=lang, slow=False)

output.save(output_filename)
```

```python
import IPython

data, sr = librosa.load(output_filename)
IPython.display.display(IPython.display.Audio(np.transpose(data), rate=sr))
```
