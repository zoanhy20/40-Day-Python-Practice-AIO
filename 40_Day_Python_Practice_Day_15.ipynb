```python
lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]

for idx, value in enumerate(lst_data):
    if value == None:
        non_none_first = -1
        for idx2 in range(idx - 1, -1, -1):
            if lst_data[idx2] != None:
                non_none_first = lst_data[idx2]
                break

        lst_data[idx] = non_none_first

lst_data
```
